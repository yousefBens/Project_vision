```python
# -*- coding: utf-8 -*-
"""

    Hyperspectral tools and demonstration for visualisation
    including direct and interactive function from the Spectral Python (SPy)  library
    and display using CMF (sRGB + Gamma)

    Status: OK/validÃ© (Nov 2023)


    Remarks : for spectral images coming from the SILIOS multispectral sensors :
              rename the file extension from bin to img

@author: nrichard
"""
import numpy as np

import scipy as sp

import spectral as spy
import spectral.io.envi as envi       ## for the spectral image load, save and plot

import matplotlib as mpl
from matplotlib import pyplot as plt  ## for the main function

##=============================================================================
def save_envi_image(Spec, filename)  :
    '''
       Save an array of spectra as an envi image for future use as groundtruth

       Arguments:
           Spec:
               an array of spectra (nblg x nbcol x nb_wavelengths) of double real

           filename:
               name of the envi spectral image file

       Return:
           img :
               the created image

    '''


    md = {'lines'  : Spec.shape[0],       ## line count
                'samples': Spec.shape[1],       ## pixel count per line
                'bands'  : Spec.shape[2],       ## wavelength count
                'data type': 5 }                ##5 =dble precision, 4= float, 12 = uint16
    #img = envi.create_image(filename+'.hdr',metadata, dtype=np.double)
    envi.save_image(filename+'.hdr', Spec, metadata = md, force=True )

    return


################################################################################
################################################################################
################################################################################
if __name__ == '__main__':
    filename ='Image_blanc_normal_00017_envi.hdr'
    path     = '/content/'

    plt.close('all') ## close all the existing figures

    ## Reading the hyperspectral image from the given file
    ## Variables to obtain :    spatial image size = (NbLg x NbCol)
    ##                          number of spectral channels : NbWaves
    ##                          array of spectral channels  :Wavelenghts[0:NbWaves]
    ##                          the spectral image : SpecImg[ 0:NbLg , 0:NbCol , 0: NbWaves]

    img = envi.open(path+filename, image = "/content/Image_blanc_normal_00017_env.img")
    #img = spy.open_image(path+filename, image="/content/Image_blanc_normal_00017_envi.img")
    NbLg, NbCol, NbWaves = img.nrows, img.ncols, img.nbands
    Wavelengths = np.array(img.bands.centers)
    SpecImg= np.array(img.load(), np.float32)     ## read the 3D array NbLg x NbCol x NbWaves


    ## Displaying image in envi mode (default channel)
    ## Colour plot of the spectral image, using the Spectral Python library
    ## Use the default bands defined in the associate text file
    plt.close('all')
    spy.imshow(img,title='cliquer sur un pixel affiche le spectre associÃ©')

    ## To access to the spectrum associated to a specific pixel location :
    lg=30
    col =10
    Spectrum = SpecImg[ lg, col , :]
    plt.figure(2)
    plt.plot(Wavelengths,Spectrum, linewidth = 3, label ="Spectrum at " + str(lg) + " x " +str(col))
    plt.xlabel("Wavelenghts (nm)")
    plt.ylabel("Radiance")
    plt.title("A spectrum")
    plt.legend()
    plt.show()

    ## To display/access to the intensity image at a specific wavelength
    channel = 2
    plt.figure(3)
    plt.imshow(SpecImg[ : , : , channel])
    plt.title("Intensity image at " + str(Wavelengths[channel]) + "nm (channel " + str(channel) + ")")



```
